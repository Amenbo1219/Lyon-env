```python
# image_size: 48*48

from google.colab import drive
drive.mount('/content/drive')

from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

from keras.utils import np_utils
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.models import load_model

from tensorflow.keras.datasets import mnist

from tensorflow.keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Flatten

from tensorflow.keras.layers import BatchNormalization

from keras.utils import np_utils
import random 
import pandas as pd


train_dir = '/content/drive/MyDrive/人工知能プロジェクト実習2/1104/CK+48_for_students/train'
valid_dir = '/content/drive/MyDrive/人工知能プロジェクト実習2/1104/CK+48_for_students/validation'
test_dir = '/content/drive/MyDrive/人工知能プロジェクト実習2/1104/CK+48_for_students/test'
# train_dir = '/content/drive/MyDrive/人工知能プロジェクト実習2/1104/CK+48_for_students_拡張前データ/train'
# valid_dir = '/content/drive/MyDrive/人工知能プロジェクト実習2/1104/CK+48_for_students_拡張前データ/validation'
# test_dir = '/content/drive/MyDrive/人工知能プロジェクト実習2/1104/CK+48_for_students_拡張前データ/test'

train_datagen = ImageDataGenerator(rescale=1./255)
valid_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(48, 48), batch_size=1000, class_mode='categorical', shuffle=True)

valid_generator = valid_datagen.flow_from_directory(
    valid_dir, target_size=(48, 48), batch_size=1000, class_mode='categorical', shuffle=True)

test_generator = test_datagen.flow_from_directory(
    test_dir, target_size=(48,48), batch_size=1000, class_mode='categorical', shuffle=True)


items = next(iter(train_generator))
x_train = items[0]
t_train_base = items[1] 
t_train_base = t_train_base.argmax(axis=1)

items = next(iter(test_generator))
x_test = items[0]
t_test_base = items[1] 
t_test_base = t_test_base.argmax(axis=1)
```

```python
l_result_train = []
l_result_test = []
for target_num in [0, 1, 2, 3, 4, 5, 6]: 
    tmp = t_train_base.copy()
    t_train = []
    for i in tmp:
        if i == target_num:
            t_train.append(1)
        else:
            t_train.append(0)

    tmp = t_test_base.copy()
    t_test = []
    for i in tmp:
        if i == target_num:
            t_test.append(1)
        else:
            t_test.append(0)

    t_train = np_utils.to_categorical(t_train)
    t_test = np_utils.to_categorical(t_test)



    ###


    # ニューラルネットワークの構築 VGG16
    model = Sequential()

    input_shape = x_train.shape[1:]
    num_classes = 2


    model = Sequential()
    model.add(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1), padding='same', input_shape=input_shape, name='block1_conv1'))
    model.add(BatchNormalization(name='bn1'))
    model.add(Activation('relu'))
    model.add(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1), padding='same', name='block1_conv2'))
    model.add(BatchNormalization(name='bn2'))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2), padding='same', name='block1_pool'))
    model.add(Conv2D(filters=128, kernel_size=(3,3), strides=(1,1), padding='same', name='block2_conv1'))
    model.add(BatchNormalization(name='bn3'))
    model.add(Activation('relu'))
    model.add(Conv2D(filters=128, kernel_size=(3,3), strides=(1,1), padding='same', name='block2_conv2'))
    model.add(BatchNormalization(name='bn4'))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2), padding='same', name='block2_pool'))
    model.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1), padding='same', name='block3_conv1'))
    model.add(BatchNormalization(name='bn5'))
    model.add(Activation('relu'))
    model.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1), padding='same', name='block3_conv2'))
    model.add(BatchNormalization(name='bn6'))
    model.add(Activation('relu'))
    model.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1), padding='same', name='block3_conv3'))
    model.add(BatchNormalization(name='bn7'))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2), padding='same', name='block3_pool'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1), padding='same', name='block4_conv1'))
    model.add(BatchNormalization(name='bn8'))
    model.add(Activation('relu'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1), padding='same', name='block4_conv2'))
    model.add(BatchNormalization(name='bn9'))
    model.add(Activation('relu'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1), padding='same', name='block4_conv3'))
    model.add(BatchNormalization(name='bn10'))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2), padding='same', name='block4_pool'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1), padding='same', name='block5_conv1'))
    model.add(BatchNormalization(name='bn11'))
    model.add(Activation('relu'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1), padding='same', name='block5_conv2'))
    model.add(BatchNormalization(name='bn12'))
    model.add(Activation('relu'))
    model.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1), padding='same', name='block5_conv3'))
    model.add(BatchNormalization(name='bn13'))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2), padding='same', name='block5_pool'))
    model.add(Flatten(name='flatten'))
    model.add(Dense(units=4096, activation='relu', name='fc1'))
    model.add(Dense(units=4096, activation='relu', name='fc2'))
    model.add(Dense(units=num_classes, activation='sigmoid', name='predictions'))


    ###


    model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])


    history = model.fit(x_train,  t_train, epochs=200, batch_size=3)

    hist_loss = history.history['loss']
    hist_acc = history.history['accuracy']

    plt.plot(np.arange(len(hist_loss)), hist_loss, label='loss')
    plt.plot(np.arange(len(hist_acc)), hist_acc, label='accuracy')
    plt.legend()
    plt.show()
    

    score = model.evaluate(x_train, t_train, verbose=0)

    # loss:score[0], accuracy:score[1]


    predictions_train1 = model.predict(x_train)
    predictions_train2 = predictions_train1.argmax(axis=1)

    predictions_test1 = model.predict(x_test)
    predictions_test2 = predictions_test1.argmax(axis=1)

    l_result_train.append([target_num, predictions_train1, predictions_train2, score[1], score[0]])
    l_result_test.append([target_num, predictions_test1, predictions_test2])

    print(target_num)
    print(score[1])
    print(predictions_test1)
    print(predictions_test2)

```

```python
def result_choice(row):
    if (sum(row)-9999) > 0:
        max_num = max(row[:7])
        result = list(row[:7]).index(max_num)
        return result
    else:
        return 9999
```

```python
def result_choice2(row):
    if list(row)[-1] == 9999:
        min_num = min(list(row)[:7])
        result = list(row)[:7].index(min_num)
        return result
    return list(row)[-1]
```

```python
def prediction_choice(x):
    max_num = max(x)
    if max_num == x[1]:
        return x[1]
    return 0
```
